# Explore Waiter Anomaly Detection Results

This notebook helps explore and visualize the results from the waiter anomaly detection pipeline.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load results
results = pd.read_parquet("../outputs/waiter_anomaly_results.parquet")
metrics = pd.read_csv("../outputs/waiter_anomaly_metrics.csv")

print(f"Total waiters: {len(results)}")
print(f"Fraud waiters: {results['is_fraud_waiter'].sum()}")
print("\nMetrics:")
print(metrics.to_string(index=False))
```

```python
# Top anomalies
score_col = [c for c in results.columns if '_score' in c][0]
top_anomalies = results.nlargest(50, score_col)

print(f"Top 50 Anomalies by {score_col}:")
print(top_anomalies[['waiter_id', 'is_fraud_waiter', score_col]].head(20))
```
